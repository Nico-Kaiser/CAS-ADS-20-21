In descriptive statistics you usually look at the summary numbers (mean, median, correlation, variance, std deviation, skewness, curtosis, etc.) and at the plots in order to get a feeling for the dataset and describe the key aspects of it.

First load the libraries / modules.

```python
# Load the needed python libraries by executing this python code (press ctrl enter)
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import pandas as pd
```

What is the difference between the median and the mean?
- the mean is sum of all values / number of values
- the median is the value in the middle of the dataset
- the mean is equal to the median if the dataset is exactly symmetrical

Calculate the skewness and the kurtosis. How are they defined?

```python
df[df['species']=='Iris-setosa'].kurt()
```

High kurtosis in a data set is an indicator that data has heavy tails or outliers. If there is a high kurtosis, then, we need to investigate why do we have so many outliers. It indicates a lot of things, maybe wrong data entry or other things. Investigate!
Low kurtosis in a data set is an indicator that data has light tails or lack of outliers. If we get low kurtosis(too good to be true), then also we need to investigate and trim the dataset of unwanted results.


```python
df[df['species']=='Iris-setosa'].skew()
```

Skewness is the degree of distortion from the symmetrical bell curve or the normal distribution. It measures the lack of symmetry in data distribution. A symmetrical distribution will have a skewness of 0.


Calculate and print correlation and covariance matrix. 

```python
df[df['species']=='Iris-setosa'].cov()
```

Covariance is a measure of the joint variability of two random variables. If the greater values of one variable mainly correspond with the greater values of the other variable, and the same holds for the lesser values, (i.e., the variables tend to show similar behavior), the covariance is positive. In the opposite case, when the greater values of one variable mainly correspond to the lesser values of the other, (i.e., the variables tend to show opposite behavior), the covariance is negative. The sign of the covariance therefore shows the tendency in the linear relationship between the variables. The magnitude of the covariance is not easy to interpret because it is not normalized.


```python
df[df['species']=='Iris-setosa'].corr()
```

What is the definition of the correlation?



Study and comment the numbers in the correlation matrix.
- length and width of the stem correlate positively (strongly) with each other
- length and width of petals only correlate very lightly (positively) with all other factors

Now we have done our descriptive statistics in numbers and tables. Now let us do it with plots. First the histograms.

```python
df_setosa = df[df['species']=='Iris-setosa']

#plt.subplot(221) #you could also do 2 smaller subplots using the code to the left, however i chose to go with full plots
df_setosa['slength'].plot(kind="hist",fill=False,histtype='step',title='Iris Setosa Setal', label="length")
ax_s = df_setosa['swidth'].plot(kind="hist",fill=False,histtype='step', label="width")
ax_s.set_xlabel('x [cm]')
ax_s.set_ylabel('A.U.')
plt.legend()
plt.show()


#plt.subplot(221)
df_setosa['plength'].plot(kind="hist",fill=False,histtype='step',title='Iris Setosa Petal', label="length")
ax_s = df_setosa['pwidth'].plot(kind="hist",fill=False,histtype='step', label="width")
ax_s.set_xlabel('x [cm]')
ax_s.set_ylabel('A.U.')
plt.legend()
plt.show()
```

Now we have studied our data with descriptive statistics. Before we can do the statistical inference, we want to make a model for our data. 

Let us start with the setal length of the Setosa species. It looks like a normal distribution so we choose that as a model. When we come to Hypothesis Testing, we will see how to test it mathematically (later in the module).

Our model will be a normal distribution with the mean and width taken from the dataset: **norm.pdf(x,mean,width)**. 

```python
from scipy.stats import norm
mean  = df_setosa['slength'].mean()
width = df_setosa['slength'].std()
print(mean,width)
# Create figure and axis
fig, ax = plt.subplots(1,1)
# Create 100 x values and plot the normal pdf for these values
#x = np.linspace(norm.ppf(0.01),norm.ppf(0.99), 100)
x = np.linspace(3,7,80)
ax.plot(x, norm.pdf(x,mean,width),'b-', lw=2, label='Normed pdf')
df_setosa['slength'].plot(kind="hist",fill=False,histtype='step',title='Iris Setosa Setal', label="length", density="True")
ax.legend(loc='best', frameon=False)
plt.show()
```

With our model we could now do a lot of inference. Taking a random leave, one could for example test how likely it is to be Iris Setosa.
